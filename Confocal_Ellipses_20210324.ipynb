```python
import numpy as np
from matplotlib import pyplot as plt

ae = 1 #distance from centre of ellipse to one focus
a = np.linspace(1.001, 1.5, 5) #semi-major axis lengths
theta = np.linspace(0, 2 * np.pi, 100)
"""
Parametric equation for an ellipse: x = a cos(theta), y = b sin(theta).
b = a * sqrt(1 - e**2) = a * sqrt(1 - (ae)**2 / a**2)
"""
plt.figure(figsize = (12, 12))
plt.plot([-ae, ae],[0, 0], 'ro', label = 'Foci')
plt.plot([0,0],[0,0], 'k*', label = 'centre')
plt.plot([0,2], [0,3], 'k--')
plt.plot([0,5], [0,0.5], 'k--')
for i in range(len(a)):
  x = a[i] * np.cos(theta)
  y = a[i] * np.sqrt(1 - (ae / a[i])**2) * np.sin(theta)
  plt.plot(x, y, label = 'a = {}'.format(np.round(a[i], decimals = 2)))
plt.xlim(-1.05 * a[-1], 1.05 * a[-1])
bmax = a[-1] * np.sqrt(1 - (ae / a[-1])**2) 
plt.ylim(-1.05 * bmax, 1.05 * bmax)
plt.legend(loc = 'best')
plt.title('Confocal ellipses')
```
